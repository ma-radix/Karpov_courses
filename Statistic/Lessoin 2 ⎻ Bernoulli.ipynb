```python
'''
1. Устройство, состоящее из пяти независимо работающих элементов, включается за время Т. 
Вероятность отказа каждого из них за это время равна 0,4. 
Найти вероятность того, что откажут три элемента. Ответ округлите до четырёх знаков после запятой. 
В качестве десятичного разделителя используйте точку.
'''
n = 5 # Общее количество элементов
k = 3 # Интересующее нас количество элементов
p = 0.4 # Вероятность интересующего нас исхода для одного элемента
q = 1-p # Вероятность НЕ интересующего нас исхода для одного элемента
C = special.comb(n, k) # Число сочетаний
P = (p**k)*(q**(n-k))*C #Общая вероятность
print(round(P, 4)) #округление выводимого значения
```

```python
'''
2. Устройство, состоящее из пяти независимо работающих элементов, включается за время Т.
Вероятность отказа каждого из них за это время равна 0,4.
Найдите вероятность того, что хотя бы один элемент откажет.
Ответ округлите до трёх знаков после запятой. В качестве десятичного разделителя используйте точку.
'''

n = 5
k = 1 # Начальная позиция
p = 0.4
q = 1-p
P = 0
# Может отказать как только одно, так и всё пять
for i in range(k, n+1):
    C = special.comb(n, i)
    P = P + (p**i)*(q**(n-i))*C
print(round(P, 3))
```

```python
'''
Для нормальной работы отдела по дата-аналитике необходимо, чтобы в офисе было не менее восьми программистов, 
а всего в штате 10 программистов. 
Вероятность того, что программист заболеет, равна 0.1. 
Найдите вероятность нормальной работы отдела по дата-аналитике.
Ответ округлите до сотых, а в качестве разделителя используйте точку.
'''

n = 10
k = 8 # Начальная позиция
p = 0.9 # Вероятность интересующего нас исхода - здоровый программист
q = 1-p
P = 0

for i in range(k, n+1):
    C = special.comb(n, i)
    P = P + (p**i)*(q**(n-i))*C
print(round(P, 2))
```

```python
'''
В очень большом чате по дата аналитике сидит 1730 человек. 
Вероятность того, что день рождения наугад взятого участника чата приходится на определенный день года, 
равна 1/365 для каждого из 365 дней. 
Найдем наиболее вероятное число счастливчиков, которые родились 1 марта.
'''

n = 1730
p = 1/365 
q = 1-p

# Наиболее вероятное число успехов: np−q <= k <= np+p
 
P1 = n*p - q
P2 = n*p + p

print(P1, "<= k <= ", P2)
# Из данного диапазона берём целое число, т.к. у нас еденица измерения - люди
```
